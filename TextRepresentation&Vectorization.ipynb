-   **URL Removal:** Eliminates URLs and hyperlinks.

    -   **Tokenization:** Breaks down text into individual words using a blank `SpaCy` model, also parallelized.
    -   **Lemmatization:** Converts words to their base forms (e.g., `running` -> `run`, `worst` -> `bad`) using `SpaCy`'s `en_core_web_sm` model, with optimized component disabling.

-   **One-Hot Encoding (OHE) (Steps 22-27):**

    -   `max_features` limits the vocabulary size. The sparse matrix is saved and loaded to demonstrate efficient handling.

    -   `CountVectorizer` with `ngram_range=(2, 2)` to extract two-word phrases.
    -   Top bigrams are displayed, and the resulting sparse matrix is saved.

-   **Bag of Trigrams (Steps 37-41):**
    -   `CountVectorizer` with `ngram_range=(3, 3)` for three-word phrases.
    -   Top trigrams are displayed, and the sparse matrix is saved.

-   **Master Vectorizer (Step 40):**
    -   A `CountVectorizer` with `ngram_range=(1, 3)` combines unigrams, bigrams, and trigrams into a single feature set (`master_matrix`) for comprehensive modeling.

-   **TF-IDF (Term Frequency-Inverse Document Frequency) (Steps 42-45):**
    -   Uses `TfidfVectorizer` with `ngram_range=(1, 2)` (unigrams and bigrams) and `sublinear_tf=True`.

## Step 6: Text Preprocessing - HTML Tag Removal
This cell uses a vectorized approach with regular expressions to remove HTML tags (e.g., `<br />`, `<div>`) from the reviews. This cleans up web-scraped text data.

````python
# Cell: Ultra-Fast Vectorized HTML Tag Removal
# Uses a non-greedy regex to find and replace all <tags> with a single space
df['clean_review_2'] = df['clean_review'].astype(str).str.replace(r'<[^<]+?>', ' ', regex=True)
````

## Step 7: Text Preprocessing - URL Removal
This cell removes URLs and hyperlinks (e.g., `http://...`, `www.`) from the reviews using regular expressions. This is crucial for cleaning text that might contain web addresses.

````python
# Cell: Ultra-Fast Vectorized URL Removal
# Catches both http/https links and naked www. domains, replacing them with a space
df['clean_review_3'] = df['clean_review_2'].astype(str).str.replace(r'http\S+|www.\S+', ' ', regex=True)
````

## Step 8: Text Preprocessing - Punctuation Removal
This cell employs a vectorized regex to remove all punctuation marks from the reviews, keeping only lowercase letters and spaces. This helps standardize the text for further analysis.

````python
# Cell: Ultra-Fast Vectorized Punctuation Removal
# Keeps only lowercase letters (a-z) and whitespaces (\s), wiping out all punctuation
df['clean_review_4'] = df['clean_review_3'].astype(str).str.replace(r'[^a-z\s]', '', regex=True)
````

## Step 9: Text Preprocessing - Text Normalization (Repeating Characters & Whitespace)
This cell performs two normalization steps:
1.  **Repeating Character Normalization:** It reduces three or more repeating characters (e.g., 'awesooomeee' to 'awesoomee') to two, preserving some emphasis while reducing noise.
2.  **Whitespace Standardization:** It collapses multiple spaces, tabs, and newlines into a single space and strips leading/trailing whitespace, ensuring consistent spacing.

````python
# Cell: Ultra-Fast Vectorized Text Normalization

# 1. Normalize Repeating Characters (Trims 3+ repeated letters down to 2)
# e.g., "awesooomeee" -> "awesoomee" (keeps words recognizable for SpaCy)
df['clean_review_5'] = df['clean_review_4'].astype(str).str.replace(r'(.)\1{2,}', r'\1\1', regex=True)

# 2. Standardize Whitespaces
# Collapses all multiple spaces, tabs, and hidden newlines into a single space
df['clean_review_6'] = df['clean_review_5'].str.replace(r'\s+', ' ', regex=True).str.strip()
````

## Step 11: Spelling Correction (SymSpellPy)
This cell initializes SymSpellPy for spelling correction. It attempts to load a frequency dictionary and then defines and applies a function `fast_spelling_correction` to correct spelling errors in the reviews. The previous run encountered an `AttributeError` during dictionary loading, which is handled here with a diagnostic message.

````python
import pandas as pd
from symspellpy import SymSpell, Verbosity
import pkg_resources
from tqdm import tqdm
import os

# 1. Initialize SymSpell and load a heavy English dictionary
# max_dictionary_edit_distance=2 allows it to catch words with up to 2 typos
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)

# Diagnostic: Check the type and attributes of sym_spell
print(f"Type of sym_spell: {type(sym_spell)}")
# print(f"Available attributes: {dir(sym_spell)}") # Uncomment if more detail is needed

try:
    dictionary_path = pkg_resources.resource_filename(
        "symspellpy", "frequency_dictionary_en_82_765.txt"
    )
    if not os.path.exists(dictionary_path):
        print(f"Warning: Dictionary file not found at {dictionary_path}. This might lead to unexpected behavior.")
    sym_spell.load_frequency_dictionary(dictionary_path, term_index=0, count_index=1)
except AttributeError as e:
    print(f"Caught AttributeError during dictionary loading: {e}")
    print("This usually means the SymSpell object is not correctly formed or the method is missing.")
    print("Please ensure symspellpy is correctly installed and imported, and try restarting the Colab runtime.")
    # As a fallback, if the dictionary loading itself is the issue, we might proceed without it
    # or use a different method, but for now, we'll just report the error.
    pass # Or handle the error as appropriate

# 2. Optimized batch loop function for 50,000 reviews
def fast_spelling_correction(dataframe_column):
    corrected_reviews = []

    # Python loops are slow; processing the strings directly via SymSpell's compound method
    # handles full sentences at once and keeps it moving at maximum speed
    for text in tqdm(dataframe_column, desc="Correcting Spelling"):
        text_str = str(text)

        # lookup_compound handles full sentences, punctuation, and splits glued words
        suggestions = sym_spell.lookup_compound(text_str, max_edit_distance=2)

        # Grab the top/best correction string
        corrected_reviews.append(suggestions[0].term if suggestions else text_str)

    return corrected_reviews

# 3. Run it across your 50,000 rows
df['clean_review_7'] = fast_spelling_correction(df['clean_review_6'])
````

## Step 13: SpaCy - Stopword Removal
This cell uses SpaCy to remove common stopwords (e.g., 'the', 'a', 'is') from the reviews. It loads a pre-trained English model, customizes the stopword list to keep negation words, and then applies a parallelized pipeline for efficient processing.

````python
import pandas as pd
import spacy
from tqdm import tqdm

# 1. Load the fast English model and disable everything except the tokenizer
# Disabling 'ner', 'parser', and 'lemmatizer' makes tokenization blazing fast
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser", "lemmatizer", "tok2vec"])

# 2. Pre-compile the stop words as a Python Set for O(1) instantaneous lookups
stop_words = nlp.Defaults.stop_words

# Prevent negation words from being stripped out during processing
words_to_keep = {"not", "no", "never", "but", "against"}
stop_words = stop_words - words_to_keep

def fast_stopword_removal(dataframe_column):
    cleaned_reviews = []

    # 3. nlp.pipe processes text as a stream in batches
    # n_process=-1 automatically utilizes ALL available CPU cores on your machine
    for doc in tqdm(nlp.pipe(dataframe_column, batch_size=2000, n_process=-1), total=len(dataframe_column)):
        # Join words that are NOT in the stop_words set and are not empty spaces
        clean_text = " ".join([token.text for token in doc if token.text not in stop_words])
        cleaned_reviews.append(clean_text)

    return cleaned_reviews

# 4. Run the parallel pipeline on your 50K rows
df['clean_review_8'] = fast_stopword_removal(df['clean_review_7'])

````

## Step 14: SpaCy - Tokenization
This cell performs tokenization using SpaCy. It initializes a blank English model (only tokenizer), which provides a very fast way to break down text into individual words or tokens. The process is parallelized across CPU cores for efficiency.

````python
import pandas as pd
import spacy
from tqdm import tqdm

# 1. Initialize a completely BLANK English model
# A blank model contains ONLY the tokenizer—no heavy neural network models are loaded!
nlp = spacy.blank("en")

def fast_tokenization(dataframe_column):
    tokenized_reviews = []

    # 2. Use nlp.pipe with n_process=-1 to run tokenization across ALL CPU cores in parallel
    # batch_size=2000 keeps memory overhead minimal while maximizing throughput
    for doc in tqdm(nlp.pipe(dataframe_column, batch_size=2000, n_process=-1), total=len(dataframe_column)):
        # Extract the text string of each token into a list
        tokens = [token.text for token in doc if token.text.strip()]
        tokenized_reviews.append(tokens)

    return tokenized_reviews

# 3. Execute the pipeline on your 50K rows
df['clean_review_9'] = fast_tokenization(df['clean_review_8'])
````

## Step 16: SpaCy - Lemmatization
This cell applies lemmatization to the tokenized reviews using SpaCy. It loads the English model, disabling unnecessary components for speed. The `fast_lemmatization` function first joins tokens into strings before passing them to `nlp.pipe` to ensure correct input for lemmatization, then extracts the base form of each word (lemma).

````python
import pandas as pd
import spacy
from tqdm import tqdm

# 1. Load the model and disable EVERYTHING except the tokenizer and lemmatizer
# Disabling 'ner' (Named Entity Recognition) and 'parser' speeds this up by 10x!
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser", "tok2vec"])

def fast_lemmatization(dataframe_column):
    lemmatized_reviews = []

    # 2. nlp.pipe processes text in batches.
    # n_process=-1 splits the workload across ALL available CPU cores automatically
    # First, join the lists of tokens into strings, then pass to nlp.pipe
    string_reviews = [" ".join(tokens) for tokens in dataframe_column]

    for doc in tqdm(nlp.pipe(string_reviews, batch_size=2000, n_process=-1), total=len(dataframe_column)):
        # Extract the lemma (base form) of each token and join them back into a clean string
        # token.lemma_ converts "boring" -> "bore", "worst" -> "bad"
        clean_text = " ".join([token.lemma_ for token in doc if token.text.strip()])
        lemmatized_reviews.append(clean_text)

    return lemmatized_reviews

# 3. Run the pipeline on your 50K reviews
df['lemmatized_review'] = fast_lemmatization(df['clean_review_9'])
````

````python
import pandas as pd

# Load your dataset (update the path to your actual file name)
df = pd.read_csv("/content/movie_reviews_lematized_review.csv")

# Enable the interactive table view
%load_ext google.colab.data_table

# Display the dataframe (this shows the interactive UI)
df.head()
````

````python
# Cell: Calculate Global Word Frequencies Efficiently
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter


# Your original Counter code will now work perfectly
word_counts = Counter([word for review in df['lemmatized_review'] for word in review])
print(f"Unique Words (Vocabulary Size): {len(word_counts):,}")

# 1. Initialize Vectorizer (binary=False to count absolute frequencies)
count_vectorizer = CountVectorizer(max_features=10000)
word_counts_matrix = count_vectorizer.fit_transform(df['lemmatized_review'])

# 2. Sum the occurrences across all rows (axis=0) for each column/word
global_counts = word_counts_matrix.sum(axis=0).A1  # .A1 flattens it to a 1D array

# 3. Map words to their total counts
vocabulary = count_vectorizer.get_feature_names_out()
word_frequencies = dict(zip(vocabulary, global_counts))

# 4. Display the top 10 most frequent words in a clean DataFrame
freq_df = pd.DataFrame(list(word_frequencies.items()), columns=['Word', 'Total_Occurrences'])
freq_df = freq_df.sort_values(by='Total_Occurrences', ascending=False).reset_index(drop=True)

print(freq_df.head(29))
````

````python
# Cell: Applying the Bag of Words Pipeline
from sklearn.feature_extraction.text import CountVectorizer

# 1. Initialize the BoW Vectorizer
# ngram_range=(1,1) means unigrams (single words).
# You can change it to (1,2) to capture word pairs like "not good" or "must watch"!
bow_vectorizer = CountVectorizer(ngram_range=(1, 1), max_features=15000)

# 2. Fit and Transform your 50k corpus into the Bag of Words matrix
bow_matrix = bow_vectorizer.fit_transform(df['lemmatized_review'])

print(f"Bag of Words Matrix Shape: {bow_matrix.shape}")
print(f"Number of non-zero elements: {bow_matrix.nnz:,}")
````

````python
# Cell: Sanity Checking the BoW Vocabulary
import pandas as pd

# 1. Grab the actual words mapped to their column positions
feature_names = bow_vectorizer.get_feature_names_out()

# 2. Sum the matrix columns to see the most frequent words caught by BoW
word_counts = bow_matrix.sum(axis=0).A1
top_words_df = pd.DataFrame({'Word': feature_names, 'Count': word_counts})
top_words_df = top_words_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

# 3. Print the top 15 words
print("Top 15 words in your Bag of Words:")
print(top_words_df.head(15))
````

````python
print("Sample Bag of Words Matrix:")
print(bow_matrix[:15, :20])
````

## Step 28: Saving the Bag of Words Sparse Matrix
Similar to the one-hot encoded matrix, the Bag of Words `bow_matrix` is also a sparse matrix. This cell saves it to a `.npz` file using `scipy.sparse.save_npz`, preserving its memory-efficient format. This allows for quick reloading for machine learning tasks.

````python
from scipy.sparse import save_npz

# Define the filename for the Bag of Words sparse matrix
bow_matrix_filename = 'bow_matrix.npz'

# Save the sparse matrix
save_npz(bow_matrix_filename, bow_matrix)

print(f"Bag of Words sparse matrix successfully saved to {bow_matrix_filename}")
````

# **Bag of Bigrams**

## Step 33: Load Data for Bag of Bigrams
This cell loads the pre-processed `lemmatized_review` data into a pandas DataFrame. It ensures a clean starting point for generating Bag of Bigrams and enables the interactive data table for easy inspection.

````python
import pandas as pd

# Load your dataset (update the path to your actual file name)
df = pd.read_csv("/content/movie_reviews_lematized_review.csv")

# Enable the interactive table view
%load_ext google.colab.data_table

# Display the dataframe (this shows the interactive UI)
df.head()
````

## Step 34: Generate Bag of Bigrams Matrix
This cell initializes and applies a `CountVectorizer` to create a Bag of Bigrams sparse matrix. `ngram_range=(2, 2)` ensures only two-word combinations are captured, and `max_features` limits the vocabulary size to the most frequent bigrams for efficiency.

````python
# Cell: High-Throughput Bag of Bigrams Generation
from sklearn.feature_extraction.text import CountVectorizer

# 1. Initialize the Vectorizer for Bigrams
# ngram_range=(2, 2) ensures ONLY word pairs are extracted
# max_features is critical here because the number of unique bigrams grows exponentially!
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), max_features=20000)

# 2. Fit and transform your 50k cleaned reviews into a Sparse Matrix
bigram_matrix = bigram_vectorizer.fit_transform(df['lemmatized_review'])

print(f"Bag of Bigrams Matrix Shape: {bigram_matrix.shape}")
print(f"Total Unique Bigrams Kept: {len(bigram_vectorizer.get_feature_names_out())}")
````

## Step 35: Analyze Top Bigrams
This cell extracts and displays the most common bigrams from the generated `bigram_matrix`. By summing the counts for each bigram across all documents, we can identify frequently co-occurring word pairs, which can provide insights into common phrases and contexts in the movie reviews.

````python
# Cell: Extract and Display Top Bigrams
import pandas as pd

# 1. Sum bigram frequencies across the corpus
bigram_counts = bigram_matrix.sum(axis=0).A1

# 2. Match bigram strings to their respective counts
bigram_features = bigram_vectorizer.get_feature_names_out()
bigram_frequencies = dict(zip(bigram_features, bigram_counts))

# 3. Sort and view the top 10 most common movie review bigrams
top_bigrams_df = pd.DataFrame(list(bigram_frequencies.items()), columns=['Bigram', 'Count'])
top_bigrams_df = top_bigrams_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

print("\nTop 10 Most Common Bigrams:")
print(top_bigrams_df.head(50))
````

## Step 36: Save Bag of Bigrams Matrix
This cell saves the generated `bigram_matrix` to disk as a sparse `.npz` file. This is crucial for long-term storage and efficient reloading, avoiding the need to re-generate the matrix for future machine learning tasks.

````python
from scipy.sparse import save_npz

# Define the filename for the Bag of Words sparse matrix
bobg_matrix_filename = 'bag_of_bigrams_matrix.npz'

# Save the sparse matrix
save_npz(bobg_matrix_filename, bigram_matrix)

print(f"Bag of Bigrams sparse matrix successfully saved to {bobg_matrix_filename}")
````

````python
import pandas as pd

# Load your dataset (update the path to your actual file name)
df = pd.read_csv("/content/movie_reviews_lematized_review.csv")

# Enable the interactive table view
%load_ext google.colab.data_table

# Display the dataframe (this shows the interactive UI)
df.head()
````

## Step 38: Generate Bag of Trigrams Matrix
This cell creates a Bag of Trigrams sparse matrix using `CountVectorizer`. `ngram_range=(3, 3)` focuses on three-word sequences, and a higher `max_features` is used to capture a broader range of these more specific phrases.

````python
# Cell: High-Throughput Bag of Trigrams Generation
from sklearn.feature_extraction.text import CountVectorizer

# 1. Initialize the Vectorizer for Trigrams
# ngram_range=(3, 3) ensures ONLY three-word combinations are extracted
# max_features=30000 filters out rare trigrams and keeps the top 30k most frequent patterns
trigram_vectorizer = CountVectorizer(ngram_range=(3, 3), max_features=30000)

# 2. Fit and transform your 50k cleaned reviews into a Sparse Matrix
trigram_matrix = trigram_vectorizer.fit_transform(df['lemmatized_review'])

print(f"Bag of Trigrams Matrix Shape: {trigram_matrix.shape}")
print(f"Total Unique Trigrams Kept: {len(trigram_vectorizer.get_feature_names_out())}")
````

## Step 39: Analyze Top Trigrams
This cell identifies and displays the most frequent trigrams from the `trigram_matrix`. Analyzing these three-word sequences can reveal more complex patterns and nuanced expressions in the review text.

````python
# Cell: Extract and Display Top Trigrams
import pandas as pd

# 1. Sum trigram frequencies across the corpus
trigram_counts = trigram_matrix.sum(axis=0).A1

# 2. Match trigram strings to their respective counts
trigram_features = trigram_vectorizer.get_feature_names_out()
trigram_frequencies = dict(zip(trigram_features, trigram_counts))

# 3. Sort and view the top 10 most common movie review trigrams
top_trigrams_df = pd.DataFrame(list(trigram_frequencies.items()), columns=['Trigram', 'Count'])
# top_trigrams_df = top_bigrams_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

print("\nTop 10 Most Common Trigrams:")
print(top_trigrams_df.head(40))
````

This cell sets up a 'master' `CountVectorizer` capable of generating a combined feature set of unigrams, bigrams, and trigrams. `ngram_range=(1, 3)` captures all three, and `max_features` limits the total number of features to 50,000, balancing expressiveness with computational feasibility. This matrix (`master_matrix`) can be used for comprehensive modeling.

````python
# The Ultimate Vectorizer Configuration for Training Classifiers
master_vectorizer = CountVectorizer(ngram_range=(1, 3), max_features=50000)
master_matrix = master_vectorizer.fit_transform(df['lemmatized_review'])
````

# **TF-IDF**

## Step 42: Load Data for TF-IDF Vectorization
This cell loads the `lemmatized_review` data again, specifically for the TF-IDF vectorization process. This ensures that the TF-IDF calculation starts with the correct pre-processed text.

````python
import pandas as pd

# Load your dataset (update the path to your actual file name)
df = pd.read_csv("/content/movie_reviews_lematized_review.csv")

# Enable the interactive table view
%load_ext google.colab.data_table

# Display the dataframe (this shows the interactive UI)
df.head()
````

## Step 43: Execute TF-IDF Vectorization Pipeline
This cell initializes and executes the TF-IDF (Term Frequency-Inverse Document Frequency) vectorization. `TfidfVectorizer` creates a sparse matrix where cell values reflect the importance of a word in a document relative to the entire corpus. `ngram_range=(1, 2)` includes unigrams and bigrams, and `sublinear_tf=True` applies a soft scaling to term frequencies.

````python
# Cell: TF-IDF Pipeline Execution
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Initialize the TF-IDF Vectorizer
# sublinear_tf=True applies a logarithmic scale to term frequency (1 + log(tf)),
# which prevents a word repeated 10 times in ONE review from overpowering the vector.
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=20000, sublinear_tf=True)

# 2. Fit and transform your 50k cleaned reviews into a Sparse TF-IDF Matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['lemmatized_review'])

# 3. Extract the full vocabulary list and their corresponding IDF scores
words = tfidf_vectorizer.get_feature_names_out()
idf_scores = tfidf_vectorizer.idf_

print(f"TF-IDF Matrix Shape: {tfidf_matrix.shape}")
print(f"Total Features Extracted: {len(words)}")
````

## Step 44: Inspect TF-IDF Vocabulary Mapping
This cell displays a sample of the vocabulary learned by the `TfidfVectorizer`. It shows how each word or bigram is mapped to a specific column index in the TF-IDF matrix, which is essential for interpreting the vector representation.

````python
# Cell: Extract Vocabulary Mapping
vocabulary_mapping = tfidf_vectorizer.vocabulary_

# Display the first 10 items from the vocabulary dictionary
print("Sample Vocabulary Mapping (Word -> Matrix Column Index):")
print(dict(list(vocabulary_mapping.items())[:10]))
````

## Step 45: Analyze IDF Scores
This cell compiles and sorts the Inverse Document Frequency (IDF) scores. IDF measures how unique or common a word is across the entire dataset. Words with low IDF scores are very common (like 'but', 'not'), while words with high IDF scores are rare and potentially more distinctive (like specific movie titles or obscure terms).

````python
# Cell: Compile IDF Scores into a Sorted DataFrame
idf_df = pd.DataFrame({
    'Word/Bigram': words,
    'IDF_Score': idf_scores
})

# Sort the dataframe to see the most common vs. most unique terms
idf_df = idf_df.sort_values(by='IDF_Score').reset_index(drop=True)

# Display the extremes
print("\n--- Top 5 Lowest IDF Scores (Most Common Words Across Corpus) ---")
print(idf_df.head(5))

print("\n--- Top 5 Highest IDF Scores (Most Unique/Rare Words Kept) ---")
print(idf_df.tail(5))
````

## Step 46: Save TF-IDF Matrix and Vocabulary
This cell saves the resulting `tfidf_matrix` to an `.npz` file and the `words` (vocabulary) to a plain text file. This allows for efficient storage and later retrieval of both the numerical representation and its corresponding features without re-running the entire TF-IDF process.

````python
from scipy.sparse import save_npz
import numpy as np

# Define filenames for the TF-IDF matrix and vocabulary
tfidf_matrix_filename = 'tfidf_matrix.npz'
tfidf_vocabulary_filename = 'tfidf_vocabulary.txt'

# Save the TF-IDF sparse matrix
save_npz(tfidf_matrix_filename, tfidf_matrix)

# Save the vocabulary (words) to a text file, one word per line
# words array is already available from previous steps (tfidf_vectorizer.get_feature_names_out())
np.savetxt(tfidf_vocabulary_filename, words, fmt='%s')

print(f"TF-IDF sparse matrix successfully saved to {tfidf_matrix_filename}")
print(f"TF-IDF vocabulary successfully saved to {tfidf_vocabulary_filename}")
````
